```
data1 = data.select(data['Code'].substr(1,1).alias("col"),"Per Capita Emissions (t) 2014")
data2 = data1.groupBy(col('col')).agg({"Per Capita Emissions (t) 2014": "avg"}).toDF("Short","CO2")
data3 = data2.where(col("CO2").isNotNull())
data3.show()
```

```
TempData = data1.join(data3, data1["col"] == data3["Short"])
display(TempData)
```

```
TempData = data1.join(data3, data1["col"] == data3["Short"], how="Left")
display(TempData)
```
